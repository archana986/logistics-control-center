# Generate Documents for Knowledge Assistant

This notebook generates markdown documents for the Knowledge Assistant to index and use for Q&A.

**Document Types Generated:**
- **Incident Analysis Reports** - Detailed analysis of specific incidents
- **Maintenance Bulletins** - Equipment maintenance advisories
- **Operational Procedures** - SOPs for incident response and rerouting
- **Customer SLA Documents** - Service level agreements by customer
- **Route Planning Guides** - Lane-specific operational guidance
- **Root Cause Analysis Reports** - Deep dive into systemic issues
- **FAA Regulatory Documents** - Compliance and regulatory context

**Output:**
Documents are written to a UC Volume for Knowledge Assistant indexing.

## Configuration

Load catalog and schema from job parameters, environment variables, or use defaults.

```python
import os


def _get_config(name: str, default: str) -> str:
    """Get config from job parameters (widgets), env vars, or default."""
    try:
        return dbutils.widgets.get(name)  # type: ignore[name-defined]
    except Exception:
        return os.environ.get(f"DATABRICKS_{name.upper()}", default)


CATALOG = _get_config("catalog", "akrishn_fe_dsa")
SCHEMA = _get_config("schema", "logistics_control_center")
VOLUME = "documents"
VOLUME_PATH = f"{CATALOG}.{SCHEMA}.{VOLUME}"

print(f"Volume path: {VOLUME_PATH}")
```

## Helper Function: Write Document to Volume

Utility function to write markdown documents to the UC Volume.

```python
def write_document_to_volume(volume_path: str, filename: str, content: str):
    """Write a document to UC Volume using dbutils."""
    try:
        volume_path_slashes = volume_path.replace('.', '/')
        full_path = f"/Volumes/{volume_path_slashes}/{filename}"
        dbutils.fs.put(full_path, content, overwrite=True)
        print(f"  ✓ Wrote {filename}")
    except Exception as e:
        print(f"  ✗ Error writing {filename}: {e}")
        raise
```
